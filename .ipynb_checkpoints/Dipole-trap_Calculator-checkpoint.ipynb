```python
#Matplotlib magic
%matplotlib inline   

import scipy as sp
import matplotlib.pyplot as plt


# The following physical constants will be useful:
#from scipy.constants import c, epsilon_0, hbar, e, k, m_e
import scipy.constants as const
# sp.c         - speed of light in vacuum
# sp.epsilon_0 - vacuum permittivity
# sp.hbar      - hbar
# sp.e         - elementary charge
# sp.k         - Boltzmann constant
# sp.m_e       - electron mass

#Specific data for 87Rb
# Rubidium mass
m_rb = 1.443e-25  #Kg

#D1 transition, 5s12_5p12 
OmegaD1 = 2*sp.pi*377.107463380e12  # Hz 794.978851156 nm
GammaD1 = 2*sp.pi*5.7500e6

#D2 transition, 5s12_5p32
OmegaD2 = 2*sp.pi*384.2304844685e12 # Hz 780.241209686 nm
GammaD2 = 2*sp.pi*6.0666e6       # Hz


def GaussBeam(z, r, w0, lamb, power):
    """ Gaussian Beam Intensity at point (z,r) for a Gaussian
    beam with given waist (w0), wavelength (lamb) and power."""
    w = waistradius(z, w0, lamb) 
    return 2*power/(sp.pi*w**2)*sp.exp(-2.0*r**2/w**2)

def waistradius(z, w0, lamb):
    """ Waist radius at longitudinal position z for a beam
    with a given waist (w0) and wavelength (lamb)."""
    zR = RayleighRadius(w0, lamb)
    return w0*sp.sqrt(1+ (z/zR)**2)
    
def RayleighRadius(w0, lamb):
    """ Rayleigh radius for a beam
    with a given waist (w0) and wavelength (lamb)."""
    return sp.pi*w0**2/lamb
    
#def I0(power, waist):
#    # Calculates intensity of light field at center of Gaussian
#    # beam.
#    return 2*power/(sp.pi*waist**2)

def U0(omega, Intensity):
    """Calculates potential depth for a given omega (2*pi*freq),
    and optical Intensity"""
    det1 = omega - OmegaD1
    det2 = omega - OmegaD2
    
    
    return (3.0/2.0)*sp.pi*const.c**2*(GammaD1/(3.0*det1*OmegaD1**3)+\
                                    2.0*GammaD2/(3.0*det2*OmegaD2**3))*\
                                    Intensity
    
    
def trapFrequencies(omega,Intensity,w0):
    """ Calculates trap frequencies for a beam with a given omega (2*pi*freq),
    intensity, waist (w0) and """
    lamb = 2*sp.pi*const.c/omega
    trapDepth = U0(omega, Intensity);
    zR = RayleighRadius(w0, lamb);
    freqx = sp.sqrt(4.0 *abs(trapDepth)/ (m_rb*w0))
    freqz = sp.sqrt(2.0 *abs(trapDepth)/ (m_rb*zR))
    return (freqx, freqz)

                    
    
def scatteringRate(omega, Intensity):
    """Calculates scattering rate for a given omega (2*pi*freq),
    and optical Intensity"""
    det1 = omega - OmegaD1
    det2 = omega - OmegaD2
    
    return (sp.pi*const.c**2/const.hbar)*(GammaD1**2/(det1**2*OmegaD1**3)+\
                                    GammaD2**2/(2.0*det2**2*OmegaD2**3))*\
                                    Intensity

    
def recoilEnergy(omega):
    k = omega/const.c
    return (const.hbar*k)**2/(2.0*m_rb)

def recoilTemperature(omega):
    return 2*recoilEnergy(omega)/const.k
    
    
def lifetimeLimit(omega, Intensity):
    return (1/scatteringRate(omega, Intensity))*\
        (-U0(omega,Intensity)/(2.0*recoilEnergy(omega)))
    
def heatingRate(omega, Intensity):
    #kappa = 1 for 3D gaussian, 0 for box potential
    kappa = 1
    return 2.0/(3.0*(1+kappa)) * recoilTemperature(omega) * scatteringRate(omega, Intensity)
    
    
    
class trapCharacteristics:
    """ Class which 'creates' a trap.
    
    Typically initialised using the wavelength, the waist and the power of the beam.
    """
    def __init__(self,lamb,waist,power):
        self.lamb = lamb;
        self.omega = 2*sp.pi*const.c/lamb
        self.waist = waist;
        self.power = power;
        self.trapDepth = -U0(self.omega,
                        GaussBeam(0.0,0.0,self.waist, self.lamb, self.power))/const.k
        self.scattRate = scatteringRate(self.omega,
                                        GaussBeam(0.0,0.0,self.waist, self.lamb, self.power))
        self.recoilTemp = recoilTemperature(self.omega)
        self.heatingRate  = heatingRate(self.omega,
                                     GaussBeam(0.0,0.0,self.waist, self.lamb, self.power))
        self.trapFreqx, self.trapFreqz = trapFrequencies(self.omega,
                                                         GaussBeam(0.0,0.0,self.waist, 
                                                                   self.lamb, self.power),
                                                         self.waist)
        self.recoilLimitedLifetime = lifetimeLimit(self.omega, 
                                                   GaussBeam(0.0,0.0,self.waist, 
                                                             self.lamb, self.power))
    
    def report(self):
        print 'Trap depth = {:.3e} K'.format(self.trapDepth)
        print 'Scattering rate = {:.3e} 1/s'.format(self.scattRate)
        print 'trap frequency x= {:.3e} Hz'.format(self.trapFreqx)
        print 'trap frequency z = {:.3e} Hz'.format(self.trapFreqz)
        print 'Recoil temperature = {:.3e} K'.format(self.recoilTemp)
        print 'Recoil limited lifetime = {:.3e} s'.format(self.recoilTemp)
        print 'Heating rate = {:.3e} K/s'.format(self.heatingRate)
        
        

        
        
```

```python
#### Dipole trap waist calculation
####
from scipy.optimize import brentq

fibre_NA = 0.12;
collimator_focal_length = 20.1 #60FC-4-M20-15
beam_diameter = 2 * collimator_focal_length * fibre_NA * 0.82

# We know the beam_diameter, and the distance to the focus (focal length) -> 
# calculate waist (in mm)
wz_diam= beam_diameter;
z = 100;
lamb = 0.000780;

#The function to be made zero is f(x) = 2*w(z) -wz_diam
beam_waist = lambda w0:wz_diam - 2*w0*sp.sqrt(1. + (z * lamb/ (sp.pi*w0**2) )**2 )   


# Need to divide by 1e3 to obtain waist in meters
waist = 1e-3*brentq(beam_waist,0.001,0.050)
waist
```

```python
## Example of trap depth for 235 mW of dipole trap power at 910nm
lamb = 782.85e-9
omega = 2*sp.pi*const.c/lamb
print 'light frequency = 2.pi.{0:.3e}'.format(omega/(2*sp.pi))

#waist = 12.5e-6
Power = 50e-3

myTrap = trapCharacteristics(lamb,waist,Power)
myTrap.report()
```

```python
## Trap depths and scattering rates for different wavelengths and powers,
## for a 10um waist trap

lambs = sp.linspace(775e-9,810e-9,200)
omegas = 2*sp.pi*const.c/lambs
#waist = 12.5e-6
Power = 5e-3
myTrap1 = trapCharacteristics(lamb,waist,Power);
myTrap1.report()
trapDepths1 = myTrap1.trapDepth;
trapDepths1.shape
scattRates1 = myTrap1.scattRate;
lifetimes1  = myTrap1.recoilLimitedLifetime;
heatRate1   = myTrap1.heatingRate;
trapFreqx1  = myTrap1.trapFreqx;
trapFreqz1  = myTrap1.trapFreqz;

#trapDepths1 = -U0(omegas, GaussBeam(0.0,0.0,waist, lambs, Power))/const.k
#scattRates1 = scatteringRate(omegas, GaussBeam(0.0,0.0,waist, lambs, Power))
#lifetimes1  = lifetimeLimit(omegas, GaussBeam(0.0,0.0,waist, lambs, Power))
#heatRate1 = heatingRate(omegas,  GaussBeam(0.0,0.0,waist, lamb, Power))
#trapFreqx1, trapFreqz1 = trapFrequencies(omegas,GaussBeam(0.0,0.0,waist, lamb, Power),waist)




Power = 10e-3
trapDepths2 = -U0(omegas, GaussBeam(0.0,0.0,waist, lambs, Power))/const.k
scattRates2 = scatteringRate(omegas, GaussBeam(0.0,0.0,waist, lambs, Power))
lifetimes2  = lifetimeLimit(omegas, GaussBeam(0.0,0.0,waist, lambs, Power))
heatRate2 = heatingRate(omegas,  GaussBeam(0.0,0.0,waist, lamb, Power))
trapFreqx2, trapFreqz2 = trapFrequencies(omegas,GaussBeam(0.0,0.0,waist, lamb, Power),waist)

Power = 50e-3
trapDepths3 = -U0(omegas, GaussBeam(0.0,0.0,waist, lambs, Power))/const.k
scattRates3 = scatteringRate(omegas, GaussBeam(0.0,0.0,waist, lambs, Power))
lifetimes3  = lifetimeLimit(omegas, GaussBeam(0.0,0.0,waist, lambs, Power))
heatRate3 = heatingRate(omegas,  GaussBeam(0.0,0.0,waist, lamb, Power))
trapFreqx3, trapFreqz3 = trapFrequencies(omegas,GaussBeam(0.0,0.0,waist, lamb, Power),waist)


plt.plot(lambs*1e9, trapDepths1,'-.', label='Power = 5mW')
plt.plot(lambs*1e9, trapDepths2,'--', label='Power = 10mW')
plt.plot(lambs*1e9, trapDepths3,'-', label='Power = 50mW')
plt.yscale('log')
plt.xlabel('Lambda [nm]')
plt.ylabel('Trap depth [K]')
plt.axvline(x=782.85, color = 'k', ls = '--')

plt.legend()
plt.savefig('trapDepth.png',dpi=300)
```
